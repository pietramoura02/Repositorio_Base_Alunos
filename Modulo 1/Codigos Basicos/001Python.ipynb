##O que são variaveis?

Variaveis são espaços na memoria do computador onde podems guardar informações (com textos, numeros, resultados de calculos, etc...) para usar depois.


variavel com texto(spring)

```python
nome = "sophia"
print (nome)
```

```python
nome = "pietra"
print (nome)
```

variavel com numero (int)

```python
idade = 15
print (idade)
```

```python
# este é um exemplo de numero inteiro negativo
temperatura = -3 # esta variavel esta armazenando um numero negativo
print(temperatura) # print() é uma função de saida
```

Variavel com numero decimal (float)

```python
altura = 1.62 # em python os numeros decimais são escritos
# escritos com ".', exemplo= 5.2 "
print (altura)
```

```python
pi =3.14
print(pi)
```

variave com valor booleano (bool)

```python
estudante = True
print (estudante)
```

## Oque é snake_case?

snake_case é um estilo usado para nome de variaveis, funções, e nomes em geral no codigo. Neste estilo, as palavras são todas em minusculas e separadas por um underline(_)

## por que usar snake_case?

*  deixa o codigo mais facil de ler
*  é o padrão recomendado pela comunidade python
*  segue a PEF8, que é  guia de estilo oficial do python

OBS: PEF8 é um documento que recomena boas práticas pra escrever código python limpo e legivel.
