# Função print()
a função print () é uma das funções mias importantes e usadas na linguagem python. sua função é, basicamente, exibir mensagens na tela ou envia-las para outro dispositivo, como imprimir dentro de um arquivo de texto.
No python 3, print() é uma função interna, de modo que não é necessario importar nenhuma biblioteca para poder utiliza-las, basta chaá-la e passar os argumentos necessarios.

```python
print("meu primeiro print") # todo texto deve estar entre aspas simples ou duplas
```

```python
print('meu segundo print')
```

```python
print("D'águas") # quando uma palavra conte aspas as aspas iniciais devem ser diferentes
```

```python
print(20+15) # o python realiza a operação que está dentro do parenteses
```

```python
print("20+15") # a operação entre aspas será apenas impresa não execultada,
#pois o python entene isso como texto
```

```python
print ("20+15 =", 20+15)
```

```python
n1= "1"
n2= "3"
print(n1 + n2) # neste exemplo o python faz a concatenação das 2 strings
```

```python
nome= "pietra"
sobrenome= "gonçalves"
nome_completo= nome + " " + sobrenome
print(nome_completo)
```
